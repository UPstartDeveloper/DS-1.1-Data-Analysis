```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Datasets/Pokemon.csv', encoding='cp1252')
df.head()
```

```python
df.describe()
```

```python
# basic Seaborn scatterplot to take a look into the distribution of ATK-DEF stats across our Pokémon!
sns.lmplot(x="Sp. Atk", y="Sp. Def", data=df)
```

```python
# Improved Scatterplot Tutorial
sns.lmplot(x="Sp. Atk", y="Sp. Def", data=df, fit_reg=False, hue='Stage')
```

```python
# MatPlotLib Customization
plt.ylim(0, None)
plt.xlim(0, None)
```

```python
# Creating a Cool Box Plot
stats_df = df.drop(["Total", "Stage", "Legendary"], axis=1)
sns.boxplot(data=stats_df)
```

```python
# Creating a Neat Violin Plot
sns.set_style("whitegrid")
sns.violinplot(x="Type 1", y="Sp. Atk", data=df)
```

```python
# Tuning our color palette for the data
pokemon_palette = ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
                   ]
sns.violinplot(x="Type 1", y="Sp. Atk", data=df, palette=pokemon_palette)
```

```python
# Creating a Swarm Plot
sns.swarmplot(x="Type 1", y="Sp. Atk", data=df, palette=pokemon_palette)
```

```python
# Creating an Overlayed Violin-Swarm Plot
plt.figure(figsize=(10,6))
sns.violinplot(x="Type 1",
               y="Sp. Atk", 
               data=df, 
               inner=None,
               palette=pokemon_palette)
sns.swarmplot(x="Type 1", 
              y="Sp. Atk", 
              data=df, 
              color='k', 
              alpha=0.7) 
plt.title("Attack by Type")
```

```python
# Creating a correlation heatmap of stats
corr = stats_df.corr()
sns.heatmap(corr)
```

```python
# Creating a histogram
sns.distplot(df.Attack)
```

```python
# Creating a Bar Plot
sns.countplot(x="Type 1", data=df, palette=pokemon_palette)
plt.xticks(rotation=-45)
```

```python
# Creating a Factor Plot
viz = sns.catplot(x="Type 1", 
                   y="Sp. Atk", 
                   data=df, 
                   hue="Stage",
                   col="Stage",
                   kind="swarm")
viz.set_xticklabels(rotation=-45)
```

```python
# Creating a Density Plot
sns.kdeplot(df.Attack, df.Defense)
```

```python
# Creating a Joint Distribution Plot
sns.jointplot(x="Sp. Atk", y="Sp. Def", data=df)
#NOTE: Add more Seaborn tutorials here! 
```
